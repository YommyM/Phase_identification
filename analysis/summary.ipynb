```python
import numpy as np
sys = 'psmdopochl'
# phase_file = '/data/gulab/yzdai/data4/phase_identification/phase_out/psmdopochl/19000-20000/psmdopochl-rawdata.xvg'/
phase_file = '/data/gulab/yzdai/data4/atomdensity/HMM/all_train/psmdopochl/train7-psmdopochl300k-rawdata.xvg'
phase_all = np.loadtxt(phase_file)[:,1:]
if 'HMM' in phase_file and (sum(phase_all[:,0] == 0) > sum(phase_all[:,0] == 1)):
    phase_all = 1 - phase_all
print(phase_all.shape)
scd_file = '/data/gulab/yzdai/data4/atomdensity/plot_data/last1us_gap5_scd_area/psmdopochl_scd.xvg'
scd_all = np.loadtxt(scd_file)[:,1:]
print(scd_all.shape)
area_file = '/data/gulab/yzdai/data4/atomdensity/plot_data/last1us_gap5_scd_area/psmdopochl_area.xvg'
area_all = np.loadtxt(area_file)[:,1:]
print(area_all.shape)

psm_id = list(range(1,181))
dopc_id = list(range(181, 325))
chol_id = list(range(325, 477))
popc_id = list(range(477, 513))
```

```python
scd_psm_ld = []; scd_psm_lo = []
scd_dopc_ld = []; scd_dopc_lo = []
scd_popc_ld = []; scd_popc_lo = []
area_psm_ld = []; area_psm_lo = []
area_dopc_ld = []; area_dopc_lo = []
area_popc_ld = []; area_popc_lo = []
area_chol_ld = []; area_chol_lo = []
ratio_psm_ld = []; ratio_psm_lo =[]
ratio_popc_ld = []; ratio_popc_lo =[]
ratio_dopc_ld = []; ratio_dopc_lo =[]
ratio_chol_ld = []; ratio_chol_lo =[]
for fr in range(phase_all.shape[0]):
    psm_ld = 0; psm_lo = 0
    dopc_ld = 0; dopc_lo = 0
    chol_ld =0; chol_lo = 0
    popc_ld =0; popc_lo = 0
    for l in range(phase_all.shape[1]):
        lip = l + 1
        phase_tag = phase_all[fr, l]
        scd = scd_all[fr, l]
        area = area_all[fr, l]
        if (lip in psm_id and phase_tag == 0):
            psm_ld += 1
            scd_psm_ld.append(scd)
            area_psm_ld.append(area)
        if (lip in psm_id and phase_tag == 1):
            psm_lo += 1
            scd_psm_lo.append(scd)
            area_psm_lo.append(area)
        if (lip in dopc_id and phase_tag == 0):
            dopc_ld += 1
            scd_dopc_ld.append(scd)
            area_dopc_ld.append(area)
        if (lip in dopc_id and phase_tag == 1):
            dopc_lo += 1
            scd_dopc_lo.append(scd)
            area_dopc_lo.append(area)
        if (lip in chol_id and phase_tag == 0):
            chol_ld += 1
            area_chol_ld.append(area)
        if (lip in chol_id and phase_tag == 1):
            chol_lo += 1
            area_chol_lo.append(area)
        if (lip in popc_id and phase_tag == 0):
            popc_ld += 1
            scd_popc_ld.append(scd)
            area_popc_ld.append(area)
        if (lip in popc_id and phase_tag == 1):
            popc_lo += 1
            scd_popc_lo.append(scd)
            area_popc_lo.append(area)
    ld_all = psm_ld + dopc_ld + chol_ld + popc_ld
    lo_all = psm_lo + dopc_lo + chol_lo + popc_lo
    ratio_psm_ld.append(psm_ld/ld_all)
    ratio_psm_lo.append(psm_lo/lo_all)
    ratio_popc_ld.append(popc_ld/ld_all)
    ratio_popc_lo.append(popc_lo/lo_all)
    ratio_dopc_ld.append(dopc_ld/ld_all)
    ratio_dopc_lo.append(dopc_lo/lo_all)
    ratio_chol_ld.append(chol_ld/ld_all)
    ratio_chol_lo.append(chol_lo/lo_all)
print(phase_file)
print('Lo---------')
print('PSM:DOPC:POPC:CHOL')
# print('%.2f:%.2f:%.2f:%.2f ' \
#       % (psm_lo/lo_all, dopc_lo/lo_all, popc_lo/lo_all, chol_lo/lo_all))
print('%.2f:%.2f:%.2f:%.2f ' \
      % (np.array(ratio_psm_lo).mean(), np.array(ratio_dopc_lo).mean(), np.array(ratio_popc_lo).mean(), np.array(ratio_chol_lo).mean()))
print('%.2f:%.2f:%.2f:%.2f ' \
      % (np.array(ratio_psm_lo).std(), np.array(ratio_dopc_lo).std(), np.array(ratio_popc_lo).std(), np.array(ratio_chol_lo).std()))
print('Scd')
print('PSM: %.2f ± %.2f' \
      % (np.array(scd_psm_lo).mean(), np.array(scd_psm_lo).std()))
print('DOPC: %.2f ± %.2f' \
      % (np.array(scd_dopc_lo).mean(), np.array(scd_dopc_lo).std()))
print('POPC: %.2f ± %.2f' \
      % (np.array(scd_popc_lo).mean(), np.array(scd_popc_lo).std()))
print('Area')
print('PSM: %.2f ± %.2f' \
      % (np.array(area_psm_lo).mean(), np.array(area_psm_lo).std()))
print('DOPC: %.2f ± %.2f' \
      % (np.array(area_dopc_lo).mean(), np.array(area_dopc_lo).std()))
print('POPC: %.2f ± %.2f' \
      % (np.array(area_popc_lo).mean(), np.array(area_popc_lo).std()))
print('CHOL: %.2f ± %.2f' \
      % (np.array(area_chol_lo).mean(), np.array(area_chol_lo).std()))
print('Ld---------')
print('PSM:DOPC:POPC:CHOL')
# print('%.2f:%.2f:%.2f:%.2f ' \
#       % (psm_ld/ld_all, dopc_ld/ld_all, popc_ld/ld_all, chol_ld/ld_all))
print('%.2f:%.2f:%.2f:%.2f ' \
      % (np.array(ratio_psm_ld).mean(), np.array(ratio_dopc_ld).mean(), np.array(ratio_popc_ld).mean(), np.array(ratio_chol_ld).mean()))
print('%.2f:%.2f:%.2f:%.2f ' \
      % (np.array(ratio_psm_ld).std(), np.array(ratio_dopc_ld).std(), np.array(ratio_popc_ld).std(), np.array(ratio_chol_ld).std()))
print('Scd')
print('PSM: %.2f ± %.2f' \
      % (np.array(scd_psm_ld).mean(), np.array(scd_psm_ld).std()))
print('DOPC: %.2f ± %.2f' \
      % (np.array(scd_dopc_ld).mean(), np.array(scd_dopc_ld).std()))
print('POPC: %.2f ± %.2f' \
      % (np.array(scd_popc_ld).mean(), np.array(scd_popc_ld).std()))
print('Area')
print('PSM: %.2f ± %.2f' \
      % (np.array(area_psm_ld).mean(), np.array(area_psm_ld).std()))
print('DOPC: %.2f ± %.2f' \
      % (np.array(area_dopc_ld).mean(), np.array(area_dopc_ld).std()))
print('POPC: %.2f ± %.2f' \
      % (np.array(area_popc_ld).mean(), np.array(area_popc_ld).std()))
print('CHOL: %.2f ± %.2f' \
      % (np.array(area_chol_ld).mean(), np.array(area_chol_ld).std()))
```

```python
def kl_divergence(mu_p, sigma_p, mu_q, sigma_q):
    return np.log(sigma_q / sigma_p) + (sigma_p**2 + (mu_p - mu_q)**2) / (2 * sigma_q**2) - 0.5
def wasserstein_distance(mu_p, sigma_p, mu_q, sigma_q):
    return np.sqrt((mu_p - mu_q)**2 + (sigma_p**2 - sigma_q**2))

kl_result = kl_divergence(np.array(scd_psm_lo).mean(), np.array(scd_psm_lo).std(),\
                          np.array(scd_psm_ld).mean(), np.array(scd_psm_ld).std())
wasserstein_result = wasserstein_distance(np.array(scd_psm_lo).mean(), np.array(scd_psm_lo).std(),\
                          np.array(scd_psm_ld).mean(), np.array(scd_psm_ld).std())
print('Scd PSM: KL: %.2f WD: %.2f' \
      %(kl_result, wasserstein_result))
kl_result = kl_divergence(np.array(scd_dopc_lo).mean(), np.array(scd_dopc_lo).std(),\
                          np.array(scd_dopc_ld).mean(), np.array(scd_dopc_ld).std())
wasserstein_result = wasserstein_distance(np.array(scd_dopc_lo).mean(), np.array(scd_dopc_lo).std(),\
                          np.array(scd_dopc_ld).mean(), np.array(scd_dopc_ld).std())
print('Scd DOPC: KL: %.2f WD: %.2f' \
      %(kl_result, wasserstein_result))
kl_result = kl_divergence(np.array(scd_popc_lo).mean(), np.array(scd_popc_lo).std(),\
                          np.array(scd_popc_ld).mean(), np.array(scd_popc_ld).std())
wasserstein_result = wasserstein_distance(np.array(scd_popc_lo).mean(), np.array(scd_popc_lo).std(),\
                          np.array(scd_popc_ld).mean(), np.array(scd_popc_ld).std())
print('Scd POPC: KL: %.2f WD: %.2f' \
      %(kl_result, wasserstein_result))


kl_result = kl_divergence(np.array(area_psm_lo).mean(), np.array(area_psm_lo).std(),\
                          np.array(area_psm_ld).mean(), np.array(area_psm_ld).std())
wasserstein_result = wasserstein_distance(np.array(area_psm_lo).mean(), np.array(area_psm_lo).std(),\
                          np.array(area_psm_ld).mean(), np.array(area_psm_ld).std())
print('Area PSM: KL: %.2f WD: %.2f' \
      %(kl_result, wasserstein_result))
kl_result = kl_divergence(np.array(area_dopc_lo).mean(), np.array(area_dopc_lo).std(),\
                          np.array(area_dopc_ld).mean(), np.array(area_dopc_ld).std())
wasserstein_result = wasserstein_distance(np.array(area_dopc_lo).mean(), np.array(area_dopc_lo).std(),\
                          np.array(area_dopc_ld).mean(), np.array(area_dopc_ld).std())
print('Area DOPC: KL: %.2f WD: %.2f' \
      %(kl_result, wasserstein_result))
kl_result = kl_divergence(np.array(area_popc_lo).mean(), np.array(area_popc_lo).std(),\
                          np.array(area_popc_ld).mean(), np.array(area_popc_ld).std())
wasserstein_result = wasserstein_distance(np.array(area_popc_lo).mean(), np.array(area_popc_lo).std(),\
                          np.array(area_popc_ld).mean(), np.array(area_popc_ld).std())
print('Area POPC: KL: %.2f WD: %.2f' \
      %(kl_result, wasserstein_result))
```

```python
def calculate_adjusted_probabilities(C_A, N_A, C_B, N_B):
	"""
	计算胆固醇出现在两类脂质中的调整后概率
	
	参数:
	C_A (int): 类别A中胆固醇的数量
	N_A (int): 类别A中脂质的总数
	C_B (int): 类别B中胆固醇的数量
	N_B (int): 类别B中脂质的总数
	
	返回:
	tuple: (P_adj_A, P_adj_B) 分别为胆固醇出现在类别A和类别B的调整后概率
	"""
	
	# 计算原始频率
	P_prime_A = C_A / N_A
	P_prime_B = C_B / N_B
	
	# 计算先验概率
	P_A = N_A / (N_A + N_B)
	P_B = N_B / (N_A + N_B)
	
	# 计算调整后的概率
	denominator = P_prime_A * P_A + P_prime_B * P_B
	P_adj_A = (P_prime_A * P_A) / denominator
	P_adj_B = 1 - P_adj_A  # 或者 (P_prime_B * P_B) / denominator
	
	# print("CHOL Lo:",P_adj_A,"CHOL Ld:",P_adj_B)
	return P_adj_A, P_adj_B

all_tags = np.array(phase_all)  # shape: (n_t, n_lipids)
N_A = np.sum(all_tags == 1)   # Lo 总数
N_B = np.sum(all_tags == 0)   # Ld 总数
lo_ratio = N_A/(N_A + N_B)
if chol_id is not None:
    # Chol 统计
    # Chol IDs 是 chol_id（如 [1, 4, 28, ...]）
    chol_ndx = np.array(chol_id) - 1 
    chol_tags = all_tags[:, chol_ndx]  # shape: (n_t, n_chol)
    C_A = np.sum(chol_tags == 1)      # chol 在 Lo 的次数
    C_B = np.sum(chol_tags == 0)      # chol 在 Ld 的次数
    p_adj_lo, p_adj_ld = calculate_adjusted_probabilities(C_A, N_A, C_B, N_B)
else: 
    p_adj_lo = None
print('lo ratio:',lo_ratio,'\nCHOL lo:',p_adj_lo)
```

```python
def calculate_adjusted_probabilities(C_A, N_A, C_B, N_B):
	"""
	计算胆固醇出现在两类脂质中的调整后概率
	
	参数:
	C_A (int): 类别A中胆固醇的数量
	N_A (int): 类别A中脂质的总数
	C_B (int): 类别B中胆固醇的数量
	N_B (int): 类别B中脂质的总数
	
	返回:
	tuple: (P_adj_A, P_adj_B) 分别为胆固醇出现在类别A和类别B的调整后概率
	"""
	
	# 计算原始频率
	P_prime_A = C_A / N_A
	P_prime_B = C_B / N_B
	
	# 计算先验概率
	P_A = N_A / (N_A + N_B)
	P_B = N_B / (N_A + N_B)
	
	# 计算调整后的概率
	denominator = P_prime_A * P_A + P_prime_B * P_B
	P_adj_A = (P_prime_A * P_A) / denominator
	P_adj_B = 1 - P_adj_A  # 或者 (P_prime_B * P_B) / denominator
	
	# print("CHOL Lo:",P_adj_A,"CHOL Ld:",P_adj_B)
	return P_adj_A, P_adj_B

all_tags = np.array(phase_all)  # shape: (n_t, n_lipids)
N_A = np.sum(all_tags == 1)   # Lo 总数
N_B = np.sum(all_tags == 0)   # Ld 总数
lo_ratio = N_A/(N_A + N_B)
if chol_id is not None:
    # Chol 统计
    # Chol IDs 是 chol_id（如 [1, 4, 28, ...]）
    chol_ndx = np.array(chol_id) - 1 
    chol_tags = all_tags[:, chol_ndx]  # shape: (n_t, n_chol)
    C_A = np.sum(chol_tags == 1)      # chol 在 Lo 的次数
    C_B = np.sum(chol_tags == 0)      # chol 在 Ld 的次数
    p_adj_lo, p_adj_ld = calculate_adjusted_probabilities(C_A, N_A, C_B, N_B)
else: 
    p_adj_lo = None
print('lo ratio:',lo_ratio,'\nCHOL lo:',p_adj_lo)
```
